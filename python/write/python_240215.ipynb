## Queue 매커니즘
1. 공백 큐 생성 : createQueue()
    이 때, front = rear = -1
2. 원소 삽입 : enQueue(A) ... enQueue(Z)
    front = -1, rear은 추가할 때 마다 1 추가(rear은 마지막 저장 위치 의미)
3. 원소 반환/삭제 : deQueue()
    front += 1 (앞 부분을 당김, 즉 마지막으로 꺼낸 자리를 의미)
이와 같은 과정을 반복하며 Queue 실행

### Queue의 상태표현
1. 초기상태 : front == rear == 1
2. 공백상태 : front == rear
3. 포화상태 : rear == n-1 (배열의 마지막 인덱스)
- del, pop(0)를 쓰면 시간이 오래 걸림
