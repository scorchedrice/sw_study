# 순차탐색
- 앞에서 부터 하나씩 찾아가는 과정으로 정렬되지 않은 데이터의 경우 많이 활용한다.

# 이진탐색
- 변수를 3개 활용한다. (시작점, 중간점, 끝점)
- **데이터가 정렬된 경우 활용 가능하다.**
- 찾으려는 데이터와 중간점 위치에 있는 데이터를 반복 비교한다. (like quick_sort)
- 재귀함수 혹은 반복문을 활용한다.
- O(log(N))

````python
# 7-2 부품찾기
N = int(input()) #가지고 있는 부품의 수
my_item_list = list(map(int, input().split()))

M = int(input()) #확인하고 싶은 부품의 수
check_item_list = list(map(int, input().split()))

# 순차적으로 확인

check_list = []
for i in check_item_list:
    no_list = []
    for j in my_item_list:
        if i == j:
            check_list += ['yes']
            continue
        else:
            no_list += ['no']
            if len(no_list) == N:
                check_list += ['no']

print(*check_list)
print('순차탐색')

# 이진탐색으로 확인
my_item_list.sort() # 정렬을 해야 활용가능하므로 sort

def binary_search(my_list, target):
    mid_index = len(my_list)//2 # 중간 index 설정
    if len(my_list) == 1:
        if my_list[0] == target: # 길이가 1일때, 확인한번 하고 참 거짓 여부 판단.
            return 'yes'
        else:
            return 'no'
    if my_list[mid_index] == target: # 길이가 1이 아닌 경우, mid_index의 값이 target과 동일하다면
        return 'yes'
    elif my_list[mid_index] > target: # target이 mid_index의 값보다 작은 경우
        return binary_search(my_list[0:mid_index], target)
    elif my_list[mid_index] <  target: # target이 mid_index의 값보다 큰 경우
        # 길이가 2인경우 mid_index가 제일 큰 값으로 잡혀, +1한 경우 indexerror이므로
        try:
            return binary_search(my_list[mid_index+1:], target)
        except IndexError:
            return 'no'


for check_item_num in check_item_list:
    print(binary_search(my_item_list, check_item_num), end = ' ')
print()
print('이진탐색')
````

````python
# 7-2 떡볶이
N, M = map(int, input().split())
ricecake_list = list(map(int, input().split()))

height_list = list(range(0,max(ricecake_list))) # 절단기 높이가 될 수 있는 리스트
# 순차탐색
def my_cut(my_list, target, height_list):
    for height_cut in height_list:
        for_result = []
        for i in range(0,len(my_list)):
            if height_cut < my_list[i]: # 절단기의 높이가 떡의 길이보다 낮아야 잘리므로
                for_result += [my_list[i] - height_cut] # 자른것들의 길이 리스트
        if sum(for_result) == target:
            return height_cut

print(my_cut(ricecake_list, M, height_list))
print('순차탐색')

# Binary
def binary_cut(my_list, target, height_list):
    mid_index = len(height_list)//2 # 높이 리스트의 중간
    
    for_result = 0
    for i in range(0,len(my_list)):
        if my_list[i] > height_list[mid_index]: # 잘리는 조건을 만족하는 경우 자른 것들의 합
            for_result += my_list[i] - height_list[mid_index]

    if for_result == target:
        return height_list[mid_index]
    elif for_result > target:
        return binary_cut(my_list, target, height_list[mid_index +1:])
    elif for_result < target:
        return binary_cut(my_list, target, height_list[0:mid_index])

print(binary_cut(ricecake_list, M, height_list))
print('순차탐색')
````
