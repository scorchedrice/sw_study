```python
# Gravity
# 상자가 낙하할 때, 떨어질 수 있는 거리는 그 숫자보다 작은 것들의 개수를 의미한다.
'''
9
7 4 2 0 0 6 0 7 0
'''
N = int(input())
box_list = list(map(int, input().split()))

count_list = []
for i in range(N):
    count = 0
    for j in range(i+1, N):
        if box_list[i] > box_list[j]:
            count += 1
    count_list += [count]

print(max(count_list))
```

```python
# baby-gin
card_list = list(range(1,10))
T = int(input())
for tc in range(1,T+1):
    my_card = list(map(int, input()))
    card_count = [0]*10
    for card_num in my_card:
        for card_type in card_list:
            if card_num == card_type:
                card_count[card_type] += 1

    run_cnt = 0
    triplet_cnt = 0
    for i in range(0,len(card_count)):
        if card_count[i] == 6: # 단 하나의 숫자로 조건을 만족하는 경우 triplet 두번
            triplet_cnt += 2
            card_count[i] = card_count[i] -6
            break
        elif card_count[i] >= 3:
            triplet_cnt += 1
            card_count[i] = card_count[i] -3

    for j in range(0,len(card_count)-2):
        if card_count[j] ==2 and card_count[j+1] == 2 and card_count[j+2] == 2:
            run_cnt += 2 # 123123처럼 두번 나오는 경우를 고려해야한다! 아니면 while 쓰던가
            break
        elif card_count[j] >=1 and card_count[j+1] >= 1 and card_count[j+2] >= 1:
            card_count[j] -= 1
            card_count[j+1] -= 1
            card_count[j+2] -= 1
            run_cnt += 1

    if triplet_cnt + run_cnt == 2:
        print(f"#{tc} Baby Gin")
    else:
        print(f"#{tc} Lose")
```

# 부분집합
- n개의 원소를 가지고 있는 집합의 부분집합의 개수는 2**n이다.

```python
# 부분집합 구하는 식
arr = [3,6,7,1,5,4]
n = len(arr)
for i in range(1<<n):
    for j in range(n):
        if i & (1<<j):
            print(arr[j], end = ", ")
    print()
print()
```

```python
# 합이 0인 부분집합들을 추출하는 방법
arr = [-7,-5,2,3,8,-2,4,6,9]

n = len(arr)
result = []

for i in range(1<<n):
    for_result = []
    count = 0
    for j in range(n):
        if i & (1<<j):
            count += arr[j]
            for_result += [arr[j]]
    if sum(for_result) == 0:
        result += [for_result]
        
print(result)
```

```python
# N개의 숫자로 구성되고 합이 K인 부분집합 개수를 구하는 방법
list_A = list(range(1,13))
n = len(list_A)
T = int(input())
for tc in range(1,T+1):
    N, K = map(int, input().split())

    result = []
    for i in range(1<<n):
        count = []
        for j in range(n):
            if i & (1<<j):
                count += [list_A[j]]
        if len(count) == N and sum(count) == K:
            result += [count]
    print(f"#{tc} {len(result)}")
```

```python
# 건물의 조망권 문제 (array를 제작하고 해결하는 방법)
for tc in range(1,11):
    N = int(input())
    building_info = list(map(int, input().split()))
    building_map = [[0]*N for _ in range(max(building_info))]

    for j in range(N):
        for i in range(building_info[j]):
            building_map[i][j] = 1

    count = 0
    for i in range(len(building_map)):
        for j in range(2,N-2):
            if building_map[i][j] != 0:
                if building_map[i][j-2] == 0 and building_map[i][j-1] == 0 and building_map[i][j+1] == 0 and building_map[i][j+2] == 0:
                    count += 1

    print(f"#{tc} {count}")
```

```python
# 건물의 조망권 문제 (건물 하나하나를 보고 이에 gravity의 방식을 적용)
for tc in range(1,11):
    N = int(input())
    building_info = list(map(int, input().split()))

    count = 0
    for i in range(2,N-2):
        height_info = building_info[i-2:i+3]
        height_info.sort()
        if height_info[-1] == building_info[i]:
            count = count + (height_info[-1]-height_info[-2])

    print(f"#{tc} {count}")
```
